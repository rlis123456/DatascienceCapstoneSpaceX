### Task 3




##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
import sqlite3
import pandas as pd

# Connect to the SQLite database
con = sqlite3.connect("Spacex.db")

# Query to calculate the total payload mass for NASA (CRS)
query = """
SELECT SUM("PAYLOAD_MASS__KG_") AS Total_Payload_Mass 
FROM SPACEXTABLE 
WHERE "Customer" = 'NASA (CRS)';
"""

# Execute the query and fetch results
result = pd.read_sql_query(query, con)
print(result)

# Close the connection
con.close()

```

### Task 4




##### Display average payload mass carried by booster version F9 v1.1


```python
import sqlite3
import pandas as pd

# Connect to the SQLite database
con = sqlite3.connect("Spacex.db")

# Query to calculate the average payload mass for booster version F9 v1.1
query = """
SELECT AVG("PAYLOAD_MASS__KG_") AS Average_Payload_Mass 
FROM SPACEXTABLE 
WHERE "Booster_Version" = 'F9 v1.1';
"""

# Execute the query and fetch results
result = pd.read_sql_query(query, con)
print(result)

# Close the connection
con.close()
```

### Task 5

##### List the date when the first succesful landing outcome in ground pad was acheived.


_Hint:Use min function_ 


```python
import sqlite3
import pandas as pd

# Connect to the SQLite database
con = sqlite3.connect("Spacex.db")

# Query to find the date of the first successful landing on a ground pad
query = """
SELECT MIN("Date") AS First_Successful_Landing_Date 
FROM SPACEXTABLE 
WHERE "Landing_Outcome" = 'True RTLS';
"""

# Execute the query and fetch results
result = pd.read_sql_query(query, con)
print(result)

# Close the connection
con.close()
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
import sqlite3
import pandas as pd

# Connect to the SQLite database
con = sqlite3.connect("Spacex.db")

# Query to find booster versions with successful drone ship landings and specific payload mass range
query = """
SELECT "Booster_Version" 
FROM SPACEXTABLE 
WHERE "Landing_Outcome" = 'True ASDS' 
AND "PAYLOAD_MASS__KG_" > 4000 
AND "PAYLOAD_MASS__KG_" < 6000;
"""

# Execute the query and fetch results
result = pd.read_sql_query(query, con)
print(result)

# Close the connection
con.close()
```

### Task 7




##### List the total number of successful and failure mission outcomes


```python
import sqlite3
import pandas as pd

# Connect to the SQLite database
con = sqlite3.connect("Spacex.db")

# Query to count successful and failed missions
query = """
SELECT 
    SUM(CASE WHEN "Mission_Outcome" LIKE 'Success%' THEN 1 ELSE 0 END) AS Successful_Missions,
    SUM(CASE WHEN "Mission_Outcome" LIKE 'Failure%' THEN 1 ELSE 0 END) AS Failed_Missions
FROM SPACEXTABLE;
"""

# Execute the query and fetch results
result = pd.read_sql_query(query, con)
print(result)

# Close the connection
con.close()
```

### Task 8



##### List the   names of the booster_versions which have carried the maximum payload mass. Use a subquery


```python
import sqlite3
import pandas as pd

# Connect to the SQLite database
con = sqlite3.connect("Spacex.db")

# Query to find booster versions that have carried the maximum payload mass
query = """
SELECT "Booster_Version" 
FROM SPACEXTABLE 
WHERE "PAYLOAD_MASS__KG_" = (SELECT MAX("PAYLOAD_MASS__KG_") FROM SPACEXTABLE);
"""

# Execute the query and fetch results
result = pd.read_sql_query(query, con)
print(result)

# Close the connection
con.close()
```

### Task 9


##### List the records which will display the month names, failure landing_outcomes in drone ship ,booster versions, launch_site for the months in year 2015.

**Note: SQLLite does not support monthnames. So you need to use  substr(Date, 6,2) as month to get the months and substr(Date,0,5)='2015' for year.**


```python
import sqlite3
import pandas as pd

# Connect to the SQLite database
con = sqlite3.connect("Spacex.db")

# Query to get the required records for failed drone ship landings in 2015
query = """
SELECT 
    SUBSTR("Date", 6, 2) AS Month, 
    "Landing_Outcome", 
    "Booster_Version", 
    "Launch_Site" 
FROM SPACEXTABLE 
WHERE "Landing_Outcome" = 'False ASDS' 
    AND SUBSTR("Date", 1, 4) = '2015';
"""

# Execute the query and fetch results
result = pd.read_sql_query(query, con)
print(result)

# Close the connection
con.close()
```

### Task 10




##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order.


```python
import pandas as pd
import sqlite3

# Define the path to the SQLite database and CSV file
db_path = "Spacex.db"
csv_path = r"C:\Users\rashm\Downloads\Spacex.csv"

# Connect to the database
con = sqlite3.connect(db_path)

# Load data from CSV into the database, creating SPACEXTABLE
df = pd.read_csv(csv_path)
df.to_sql("SPACEXTABLE", con, if_exists="replace", index=False)

# Close the connection to ensure the table is saved
con.close()

```

```python
# Reconnect to the database
con = sqlite3.connect(db_path)

# Define the query to rank landing outcomes by count
query = """
SELECT 
    "Landing_Outcome", 
    COUNT(*) AS Outcome_Count 
FROM SPACEXTABLE 
WHERE "Date" BETWEEN '2010-06-04' AND '2017-03-20' 
GROUP BY "Landing_Outcome" 
ORDER BY Outcome_Count DESC;
"""

# Execute the query and fetch results
result = pd.read_sql_query(query, con)
print(result)

# Close the connection
con.close()
```
